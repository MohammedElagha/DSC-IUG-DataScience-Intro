# Data Visualization

### Data

```
covid_2020 = [ ('Jan', 15), ('Feb', 250), ('Mar', 3454), ('Apr', 12401), ('May', 35641), ('Jun', 80641), 
              ('Jul', 100141), ('Aug', 108754), ('Sep', 142477), ('Oct', 180454), ('Nov', 190545), ('Dec', 200465) ]
```

```
import pandas as pd
```

```
df = pd.DataFrame(data=covid_2020)
```

```
df
```

```
df.columns = ['Month', 'Cases_no']
```

```
df
```



---



```
import matplotlib.pyplot as plt
```

### Scatter Plot

```
# create a figure and axis
fig, ax = plt.subplots()

# scatter the Month against the Cases_no
ax.scatter(df['Month'], df['Cases_no'])

# set a title and labels
ax.set_title('COVID-19 Cases for 2020')
ax.set_xlabel('Month')
ax.set_ylabel('Cases no.')
```

### Line Plot

```
# create a figure and axis
fig, ax = plt.subplots()

# scatter the Month against the Cases_no
ax.plot(df['Month'], df['Cases_no'])

# set a title and labels
ax.set_title('COVID-19 Cases for 2020')
ax.set_xlabel('Month')
ax.set_ylabel('Cases no.')
```

### Bar Chart

```
# create a figure and axis
fig, ax = plt.subplots()

# scatter the Month against the Cases_no
ax.bar(df['Month'], df['Cases_no'])

# set a title and labels
ax.set_title('COVID-19 Cases for 2020')
ax.set_xlabel('Month')
ax.set_ylabel('Cases no.')
```
